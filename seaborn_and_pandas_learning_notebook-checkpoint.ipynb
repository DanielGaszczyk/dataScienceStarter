```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
sns.set_theme()
df = sns.load_dataset("penguins")
sns.pairplot(df,hue="species")
```

```python
#sns.color_palette("rocket", as_cmap=True)
sns.histplot(df,x="island",y="body_mass_g", color="pink")
sns.displot(df,x="body_mass_g",color="orange")
```
